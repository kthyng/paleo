```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0,10)

fig = plt.figure()
ax = fig.add_subplot(111)
ax.plot(x, x)

# overlay axes
ax2 = fig.add_axes([0.2, .7, .1, .1])
im = plt.imread('https://upload.wikimedia.org/wikipedia/commons/4/47/PNG_transparency_demonstration_1.png')
ax2.imshow(im)
```
